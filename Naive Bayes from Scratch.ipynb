```python
# Importing Necessary modules

import sklearn 
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,f1_score
from math import e,pi,exp,sqrt
```

```python
# Loading the data

df=pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv')
df.columns=list('ABCDEFGHI')
df
```

Data is already cleaned, so we can directly give it to the model. We can also do preprocessing, to churn out accuracy

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop('I',axis=1),df["I"],test_size=0.2)
```

```python
# Extracting the values

x_train=x_train.values
x_test=x_test.values
y_train=y_train.values
y_test=y_test.values
```

```python
# Naive Bayes Class
class NaiveBayes:
    
    def summarize(self,x,y):
        s=set(y)
        d={}
        for i in range(len(y)):
            if(y[i] in d):
                d[y[i]].append(x[i])
            else:
                d[y[i]]=[x[i]]
        return d
    
    def mean(self,l):
        return sum(l)/len(l)
    
    def std(self,l):
        m=self.mean(l)
        l1=[(i-m)**2 for i in l]
        return (sum(l1)/len(l1))**0.5
    
    def calc(self,d):
        d1={}
        for i in d:
            d1[i]=[]
        for i in d:
            for j in zip(*d[i]):
                m=self.mean(j)
                s=self.std(j)
                d1[i].append([m,s])
        return d1
    
    def probability(self,x,m,s):
        exp1=e**((((x-m)/s)**2)*0.5)
        return (1/(((2*pi)**0.5)*s))*exp1
    
    def probability1(self,x, mean, stdev):
        exponent = exp(-((x-mean)*2 / (2 * stdev*2 )))
        return (1 / (sqrt(2 * pi) * stdev)) * exponent
        
    def fit(self,x,y):
        self.d=self.summarize(x,y)
        self.d1=self.calc(self.d)
        self.l=[]
        for i in self.d:
            self.l.append(len(self.d[i]))
    
    def predict(self,x):
        res=[]
        c=x.shape[1]
        for i in x:
            p=[]
            for j in self.d1:
                c1=self.l[j]
                for k in range(c):
                    c1*=self.probability1(i[k],self.d1[j][k][0],self.d1[j][k][1])
                p.append([j,c1])
            p.sort(key=lambda val:val[1])
            res.append(p[-1][0])
        return res     
```

```python
model=NaiveBayes()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
# Accuracy of the custom coded model on the test-set

print('Accuracy is: ',accuracy_score(y_test,y_pred))
```

```python
# F1-Score of the custom coded model on the test-set

print('F1-Score is: ',f1_score(y_test,y_pred,average='weighted'))
```
